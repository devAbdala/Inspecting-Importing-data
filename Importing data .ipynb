Web Scraping

```python
url = "http://www.wikipedia.org/"
request = Request(url)
response = urlopen(request)
data = response.read()
data
```

```python
response.close()
```

```python
url = "http://www.wikipedia.org/"
response = requests.get(url)
html = response.text
html
```

```python
output = BeautifulSoup(html)
print(output)
```

```python
print(output.prettify())
```

```python
print(output.title)
```

```python
print(output.get_text())
```

```python
for link in output.find_all('a'):
    print(link.get('href'))
```

```python
for div in output.find_all('meta'):
    print(div.get('content'))
```
