```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
# universe Resource locator
url = "https://wuzzuf.net/search/jobs/?a=hpb%7Cspbg&filters%5Bcity%5D%5B0%5D=Cairo&filters%5Bcountry%5D%5B0%5D=Egypt&q=data%20analyst"
response = requests.get(url)
```

```python
html = response.text # Getting html code of page
```

```python
wuzzuf = BeautifulSoup(html , "html.parser")
```

```python
wuzzuf
```

```python
containers = wuzzuf.find_all("div" , {'class' : "css-1gatmva e1v1l3u10" })
len(containers)
```

```python
BeautifulSoup.prettify(containers[0])
```

```python
containers[0].div.h2.text
```

```python
job_title = containers[0].findAll("h2" , {"class" : "css-m604qf"})
job_title[0].text
```

```python
company_name = containers[0].findAll("a" , {"class" : "css-17s97q8"})
company_name[0].text
```

```python
time = containers[0].findAll("span" , {"class" : "css-1ve4b75 eoyjyou0"})
time[0].text
```

```python
file = open("./datasets/wuzzuf.csv" , mode = 'w')
header = "job_title,company_name,job_type\n"
file.write(header)
```

```python
for container in containers :
    job_title = container.findAll("h2" , {"class" : "css-m604qf"})
    job_title = job_title[0].text.strip()
    
    company_name = container.findAll("a" , {"class" : "css-17s97q8"})
    company_name = company_name[0].text.strip()
    
    time = container.findAll("span" , {"class" : "css-1ve4b75 eoyjyou0"})
    time = time[0].text.strip()
    
  #  print(job_title)
  #  print(company_name)
  #  print(time)
  #  print()
    
 #   print(job_title  +  ", "  +  company_name  +  ", "  +  time)
 #   print()
    file.write(job_title  +  ", "  +  company_name  +  ", "  +  time + "\n")
```

```python
file.close()
```
